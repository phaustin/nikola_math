# Density and pressure in the ocean and atmosphere

## 1. Units

Units of force = mass x acceleration = Newtons = $kg\, m\,s^{-2}$

Units of pressure = force/area = Pascals = $N\,m^{-2}$.  Note that atmospheric pressure is usually reported
in hPa (1 hectoPascal = 100 Pa) or kPa (1 kiloPascal = 1000 Pa)

## 2. Hydrostatic balance

The ocean or atmosphere is in **hydrostatic balance** when the pressure force holding
it up is equal to the gravitational force dragging it down:

```python
Image('hydrostatic.png',width='30%')
```

The figure shows a 1 $m^{2}    $ section with constant density $\rho$ ($kg\,m^{-3}$) and total mass for the 1 $m^2$ = $\rho dz$ ($kg\,m^{-2}$).  If the layer is not going to move up or down,
then the pressure difference has to be equal to the gravitation force:

$$dp = - \rho g dz$$

## Pressure with depth in the ocean

Liquid water is less compressible than steel, so we can assume the density $\rho = 1000\ kg\,m^{-3}$ is independent of depth.  At the 
surface (z=0) we know that the pressure is $p_0$ =1 atmosphere = 100 kPa.  What is the pressure 10 meters below the surface?
Integrating from z= -10 meters, press=$p_0$ to press = p:

$$\int_{p_0}^p dp^\prime = -\int_{-10}^0 \rho g dz^\prime$$

$$p - p_0 = -\rho g (0 - (-10)) $$

$$p = p_0 + \rho g (10) = 100,000\ Pa + 1000*9.8*10 = 198\ kPa=\text{2 atmospheres}$$

This is why scuba divers show up in hospital emergency rooms -- rapidly rising from 20 meters below the surface means
the body undergoes a pressure change from 3 atmospheres to 1 atmosphere, creating non-equilbrium nitrogen
gas concentrations  in the bloodstream and causing the bends.

**Important point:**  the $mass/m^2$ between any two heights in the ocean or atmosphere is just the pressure difference across those heights divided by $g$, that is:

Define $M$, the $mass/m^2$ in a layer as:

$$M=\int_{z_1}^{z_2} \rho dz $$

From the hydrostatic equation:

$$\int_{p_1}^{p_2} dp = (p_2 - p_1) = -g \int_{z_1}^{z_2} \rho dz = -g M $$

and if we write $\Delta p = p_1 - p_2$, which is positive  since pressure decreases with increasing $z$ in the ocean
and atmosphere:

$$M = \frac{\Delta p}{g}$$

## Pressure with height in the atmosphere

The atmosphere is definitely compressible, and so the density will vary with height.  I'll show in a separate
article that for a hydrostatic atmosphere the density dependence is approximately:

$$\rho(z) = \rho_0 \exp(-z/H)$$

where $H$, called the **density scale height**, varies with latitude but is roughly 8 km, and $\rho_0$ is
related to the surface pressure and temperature by the ideal gas law:

$$\rho_0 = \frac{p_0}{R_d T_0}$$

where $R_d = 287\ J\,kg^{-1}\,K^{-1}$ is the gas constant for dry air.

We could integrate $\rho(z) g dz$ to get the pressure as a function of height, but if all we need is the mass
between two pressure levels, we don't need to, this equation still works fine:

$$M = \frac{\Delta p}{g}$$

Putting in some numbers, if $\Delta p$ = 100 kPa (the entire atmosphere) then $M=10,000\ kg\,m^{-3}$, i.e.
the mass of the air above us at sea level in a 1 $m^2$ column is 10 metric tonnes.   This is why
things like this happen: [tanker car](https://www.youtube.com/watch?v=2WJVHtF8GwI&list=PL8UshAqCEEvBPUtj6bThjh20bMhvNmeME&index=3)
