# Working with historgrams: Introduction

The radiative flux from a blackbody depends on wavelength/wavenumber and temperature.  That dependence is described by the Planck function, which gives the distribution (histogram) of fluxes for different wavenumbers at a particular blackbody temperature.  This page makes the connection between the Planck curve and the histogram of class grades.

##  1. Example -- class grade distribution

 The following figure shows a grade distribution for 1000 students with an average of 75% following a bell curve:

```python
Image(filename='grade_distribution.png',width='40%')
```

### Figure 1: Grade distribution with uniform bins

In the figure above, there are 20 bins, each with a width of 5% to cover the mark range from 0-100%

But what if we want to use UBC's marking scheme, with A+ = 90-100, A= 85-90, etc.   Now the bin widths are not uniform, and the same data looks like this:


```python
Image(filename='grade_distribution_UBC.png',width='40%')
```

### Figure 2:  Grade distribution with UBC bins

To reiterate, this is exactly the same data, just binned with UBC grade boundaries.  We've got 111 of 1000 students who scored below 50%, and 113 who scored above 90%.

To get a better feel for the distribution, we need to remove the effect of the different bin widths by dividing the counts in each bin by the width of the bins:

```python
Image('grade_distribution_UBC_density.png',width='40%')
```

### Figure 3: Grade distribution histogram in density units

Comparing the three figures, we can see that in Figures 1 and 3, the area of each bin is proportional to the number of students in the bin.  For example 113 of 1000 students scored 90 or higher, and the height of the A+ bin in Figure 3 is 11.3, with a bin width of 10%.   In Figure 2 it's the bin height that is 113, and the bin area is meaningless.

## 2. Planck function as a histogram

Suppose I have a camera that measures the flux $I$ in 20 different wavnumber ranges from 0 to 2500 $cm^{-1}$.  If I point that camera at a 300 degree blackbody I would get readings that looked something like this:

```python
Image('planck_bins.png',width='40%')
```

### Figure 4.   Observed flux from 300 K blackbody

Figure 4 has the same problem as Figures 1 and 2 above.  The total area doesn't have any physical meaning, and if we had wavenumber bins of different widths we'd get a distorted picture of how much flux was in each wavenuber range.

```python
Image('planck_density.png',width='40%')
```

### Figure 5.  Observed flux, area of histogram = total flux

Figure 5 shows the same data with the flux in each bin divided by the bin width of 125 $cm^{-1}$.  Now the area of the histogram is equal to the total flux (460 $W\,m^{-2}$) and we could use wavenumber ranges of different widths without distortion if we wanted to.

# Summary

The Planck function has units of $W\,m^{-2}\ per\ cm^{-1}$ or $W\,m^{-2}\ per\ \mu m$.  This allows us to interpret the area between two wavenumber intervals as the flux arriving at the instrument for photons in that wavenumber range.
