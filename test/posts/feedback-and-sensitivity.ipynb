### 1. Relate climate feedback f, climate sensitivity Î», and column energy E

In the day 10 reading we introduced the following equation:

$$\underbrace{\frac{dE}{dt}}_{\text{layer heating}} = \underbrace{\Delta F}_{\text{forcing}} + \underbrace{\Delta R}_{\text{response}}$$

and in Day 9 slide 6 we showed what the response $\Delta R$ looked like in terms of
$f_{planck}$, the planck feedback.  Specifically, start with a system that
is in equilibrium  so that $dE/dt = 0$. We know that the ground temperature
is 243.7 K so that $I_g$ is -200 $W\,m^{-2}$

```python
Tg = (200/sigma)**0.25
print('Tg = {:5.2f}'.format(Tg))
Image('equil_200.png',width='30%')
```

Now suppose $I_0$ increases to 202 $W\,m^{-2}$.  The new situation is:

```python
Image('non_equil_202.png',width='30%')
```

So the heating rate for the layer is now:

$$\underbrace{\frac{dE}{dt}}_{\text{layer heating}} = \underbrace{\Delta F}_{\text{forcing}} + \underbrace{\Delta R}_{\text{response}} = \Delta F -4\sigma T^3 \Delta T = \Delta F + f_{planck} \Delta T$$

where $f_{planck}$ is -3.28 $W\,m^{-2} K^{-2}$:

```python
f_planck = -4*sigma*Tg**3.
print('Planck feedback is {:5.2f} W/m^2/K'.format(f_planck))
```

What is the climate sensitivity for the layer?  From the day 10 reading equation 5 we
know that 

$$\frac{dE}{dt} = \Delta F - \frac{\Delta T}{\lambda}$$

so comparing these two equations gives $\lambda = -\frac{1}{f_{planck}} = -1/(-3.28) = 0.3\ K/(W\,m^{-2})$

### Section 1 Summary

The smaller the stabilizing $\text{(negative)}$ feedback, the larger the climate sensitivity.  That
agrees with the common use of the word sensitivity -- sensitive systems without stabilizing
feedbacks are easily thrown out of equilibrium.

### 2. E and $h_m$ for an atmospheric layer

How do we calculate E, the energy in a 1 $m^2$ column, if we know $h_m$, the moist static
energy defined in Day 9?   First recognize that the $J\,m^{-2}$ in a thin layer of
thickness dz is given by:

$$\Delta E = \rho h_m \Delta z$$

Units check:  $J\,m^{-2} = kg\,m^{-3} \times J\,kg^{-1} \times m$

For thicker layers, we just need to integrate, and since we showed on Day 9 that $h_m$ is
roughly constant with height, it can come out of the integral:

$$E = \int_{bot}^{top} dE =\int_{bot}^{top} \rho h_m dz = h_m \int_{bot}^{top} \rho dz = M h_m$$

where $M = \int_{bot}^{top} \rho dz $ is just the mass in a ${m^2}$ column.

How do we get a number for $M$?  From the [density and pressure notes](http://clouds.eos.ubc.ca/~phil/courses/e340review/posts/density-and-pressure/) we know that if the pressure drop across the layer is ${\Delta p}$ (in Pascals) then:

$$M = \frac{\Delta p}{g} $$

Question:  what is $M$ for the layer between 100 kPa (surface) and 90 kPa (about 1 km up). 
(Answer: about a metric tonne/$m^2$)

```python
M = 1000*(100 - 90)/9.8
print('M = {:5.2f} kg/m^2'.format(M))
```

### 3. E and $h_m$ for an ocean layer

The ocean overturns very slowly and doesn't have a gas phase, so there is no potential or
latent heat term, and the moist static energy reduces to $h=c_p T$, where $c_p$, the heat
capacity for liquid water, is about 4218 $J\,kg^{-1}\,K^{-1}$.  If we write $\rho_w \approx$ 1020 $kg/m^{-3}$ for the density of seawater, then the column energy of an ocean layer of depth $H$ is


$$E = \int_{bot}^{top} dE =\int_{bot}^{top} \rho_w h dz = \rho_w c_p T \int_{bot}^{top}dz = \rho_w H c_p T$$

Putting this all together, we've got

$$\frac{dE}{dt} = \rho_w H c_p \frac{dT}{dt} = \Delta F - \frac{\Delta T}{\lambda}= \Delta F + f_{planck} \Delta T$$

As we'll see after the break, we can solve this equation to get a climate forecast, which is the temperature
as function of time given a constant forcing $\Delta F$
