# Planck feedback

In the Day 8 worksheet, we found that we would need to increase the surface
temperature by about 9 degrees to counteract $\Delta F$ = 43 $W\,m^{-2}$ of greenhouse trapping. This $\Delta I$ cooling response is called the Planck feedback -- in this case, for every degree you raise the surface temperature, you make the upward flux $I$ about -43/9= -4.8 $W\,m^{-2}$ more negative.  This is a stabilizing
feedback because sooner or later the black surface  will get hot enough to generate a negative $\Delta I$ that can counteract any amount of positive forcing $\Delta F$.
Also note that it works in the other direction -- if $\Delta F$ is negative then the ground temperature will drop and $\Delta I$ will become less negative = more positive. 

## Calculating the Planck feedback

So how would we calculate this without Modtran?  Start with an example where the fluxes are
in balance at 400 $W\,m^{-2}$ and add 2 $W\,m^{-2}$:



### Suppose a black surface which is emitting 400 $W\,m^{-2}$ in equilibrium receives a downward flux of 402 $W\,m^{-2}$.  What is its new equilibrium temperture?

```python
Image('surface_402.png',width='30%')
```

### 1) Solve by using the Stefan Boltzman equation

```python
old_flux = 400.
T_ground = (old_flux/sigma)**(1/4.)
new_flux = 402.
T_ground_new = (new_flux/sigma)**(1./4.)
print('old T_ground={:6.3f} K'.format(T_ground))
print('new T_ground={:6.3f} K'.format(T_ground_new))
print('difference: {:6.3f} K'.format(T_ground_new - T_ground))
```

So the Planck feeback is -2 $W\,m^{-2}$/0.362 degrees  = -5.52 $W\,m^{-2} K^{-1}$

```python
f_planck= -2/0.362
print('f_planck={:5.2f} W/m^{{-2}}/K'.format(f_planck))
```

### 2) Solve using derivatives

Graphically we can find the $\Delta T$ we need to produce a particular $\Delta I$ using the
slope of the Stefan-Boltzman equation:

```python
Image('stefan_boltzman_deriv.png',width='30%')
```

In math notation: the temperature change we need to get 2 $W\,m^{-2}$ of extra
emission using the derivative of the Stefan-Boltzman equation:

$$\Delta I_g = \frac{dI_g}{dT} \Delta T = \frac{d (-\sigma T^4)}{dT} \Delta T = -4 \sigma T^3 \Delta T = f_{planck} \Delta T$$



```python
Image('stefan_boltzman_deriv.png',width='50%')
```

$\Delta I_g = f_{planck} \Delta T_g = -4 \sigma T_g^3 \Delta T_g = -5.52\ W\,m^{-2} K^{-1} \times \Delta T_g$

```python
f_planck = -4*sigma*T_ground**3.
print('approximate planck feedback is {:5.2f} W/m^{{2}}/K'.format(f_planck))
```
